```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
avg_speed_mad4qn = np.load('/content/avg_speed_mad4qn_0.npy')
avg_speed_atl1= np.load('/content/avg_speed_atl1_0.npy')
avg_speed_fttl1=np.load('/content/avg_speed_fttl1_0.npy')

```

# Avg_speed comparison

```python
import numpy as np
import matplotlib.pyplot as plt

avg_speed_histo2 = {
    'FTTL1': np.mean(avg_speed_fttl1),
    'ATL1': np.mean(avg_speed_atl1),
    'MAD4QN-PS': np.mean(avg_speed_mad4qn)
}

x = range(3)
bar_width = 0.65

facecolors = ['green', 'chocolate', 'gray', 'red']
edgecolors = facecolors

for i, pair in enumerate(avg_speed_histo2.values()):
    plt.bar(x[i], pair, bar_width, color=facecolors[i % len(facecolors)], edgecolor=edgecolors[i % len(edgecolors)], alpha=0.8)

plt.xticks(range(3), list(avg_speed_histo2.keys()), fontsize=14)
plt.ylabel('Speed [m/s]', fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y')
plt.savefig("avg_speed.png", format='png')
plt.show()
```

# Avg_travel_time vs waiting_time

```python
travel_time_fttl1 = np.load('/content/travel_time_fttl1_0.npy')


waiting_time_fttl1 = np.load('/content/waiting_time_fttl1_0.npy')




# ATL 1

travel_time_atl1 = np.load('/content/travel_time_atl1_0.npy')

waiting_time_atl1 = np.load('/content/waiting_time_atl1_0.npy')

# MAD4QN-PS

travel_time_mad4qn = np.load('/content/travel_time_mad4qn_0.npy')

waiting_time_mad4qn = np.load('/content/waiting_time_mad4qn_0.npy')



#collisions_mad4qn = np.load('data/collisions_mad4qn_0.npy')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming travel_time_fttl1, waiting_time_fttl1, etc. are defined

travel_time_histo2 = {
    'FTTL1': [np.mean(travel_time_fttl1), np.mean(waiting_time_fttl1)],
    'ATL1': [np.mean(travel_time_atl1), np.mean(waiting_time_atl1)],
    'MAD4QN-PS': [np.mean(travel_time_mad4qn), np.mean(waiting_time_mad4qn)]
}

x = range(3)  # Changed from range(0, 6) to range(3)
bar_width = 0.4

facecolors = ['green', 'darkblue', 'gray', 'red']
edgecolors = facecolors

for i, pair in enumerate(travel_time_histo2.values()):
    plt.bar(x[i] - bar_width/2, pair[0], bar_width, color=facecolors[i % len(facecolors)], edgecolor=edgecolors[i % len(edgecolors)], alpha=0.8, label='Average travel time per vehicle' if i == 0 else "")
    plt.bar(x[i] + bar_width/2, pair[1], bar_width, color=facecolors[i % len(facecolors)], edgecolor=edgecolors[i % len(edgecolors)], alpha=0.3, label='Average waiting time per vehicle' if i == 0 else "")

plt.xticks(range(3), list(travel_time_histo2.keys()), fontsize=14)  # Changed from range(0, 6) to range(3)
plt.ylabel('Time [s]', fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='upper right', fontsize=12)
plt.grid(axis='y')
plt.savefig("travel_time.pdf", format='pdf')
plt.show()
```
